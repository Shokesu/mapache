```python
import numpy as np

%matplotlib inline
```

```python
import sys
sys.path.append('../')

import mapache
```

```python
spain_parties = mapache.PartySet()
```

```python
ciudadanos = mapache.Party('Ciudadanos',
                   'https://www.ciudadanos-cs.org/var/public/sections/page-imagen-del-partido/logo-ciudadanos.jpg',
                   short_name = 'C\'s',
                   full_name = 'Ciudadanos - Partido de la Ciudadanía')

spain_parties.add(ciudadanos)
```

```python
ciudadanos.show()
```

```python
psoe = mapache.Party('PSOE', 'http://psoesoria.net/wp-content/uploads/2015/11/nuevo-logo-psoe-01.png',
                   short_name = 'PSOE',
                   full_name = 'Partido Socialista Obrero Español')

spain_parties.add(psoe)
```

```python
pp = mapache.Party('PP', 'http://cdn.20m.es/img2/recortes/2015/07/09/233009-390-389.jpg',
                   short_name = 'PP',
                   full_name = 'Partido Popular')

spain_parties.add(pp)
```

```python
podemos = mapache.Party('Podemos', 'https://podemosespinar.files.wordpress.com/2014/07/logopodemosdos.jpg',
                   short_name = 'POD',
                   full_name = 'Podemos')

spain_parties.add(podemos)
```

```python
polls = {
        'title': 'Encuestas',
        'dates': np.asarray([0,0.3,1,1.3,2,2.5,3,4,4.1,4.2]).reshape(-1,1),
        'parties' : {
                    'ciudadanos' : np.asarray([10,13,12,19,18,20,21,22,23,26]),
                    'podemos' : np.asarray([8,10,9,10,12,15,14,13,14,14]),
                    'psoe' : np.asarray([22,21,22,20,23,22,21,19,18,17]),
                    'pp' : np.asarray([32,31,31,29,30,31,30,31,31,32])
                 }
        }


elections = {
        'title': '20D',
        'dates': np.asarray([6]).reshape(-1,1),
        'parties' : {

            'ciudadanos' : np.asarray([25]),
            'podemos' : np.asarray([16]),
            'psoe' : np.asarray([18]),
            'pp' : np.asarray([30.5])
            }
    }
        
```

```python
pv = mapache.PollVis(spain_parties)

# Multiple polls column
pv.add_column(polls)

# Single poll column (ie. election results)
pv.add_column(elections)
```
